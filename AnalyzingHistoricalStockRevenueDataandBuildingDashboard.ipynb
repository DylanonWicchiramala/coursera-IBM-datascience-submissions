## Question 2: Use Webscraping to Extract Tesla Revenue Data

```python
def get_financial_data(ticker:str):
    """ Get financial data from stock ticker.
    return pd.DataFrame of financial data
    """
    # url to tesla financial data.
    url = f"https://finance.yahoo.com/quote/{ticker}/financials?p={ticker}"
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'}


    res = requests.get(url, headers=headers)

    soup = bs.BeautifulSoup(res.content, 'html5lib')

    # get time series.
    column_name = [column_name.text for column_name in soup.find('div', {'class':'D(tbr) C($primaryColor)'})][1:]

    financial_data = pd.DataFrame(columns=column_name)

    # get data.
    for i, row in enumerate(soup.find("div", {'class':'D(tbrg)'}).find_all('div',{'class':'D(tbr) fi-row Bgc($hoverBgColor):h'})):
        cols = [e.text for e in row.find_all('div')]
        data = pd.DataFrame({cols[0]: cols[3:]}, index=column_name).T
        financial_data = pd.concat([financial_data, data])

    return financial_data.T

financial_data = get_financial_data
tesla_revenue = financial_data('tsla')[['Total Revenue']]
tesla_revenue.tail()
```

## Question 5: Plot Tesla Stock Graph

```python
import plotly.graph_objects as go

import pandas as pd
from datetime import datetime

def make_graph(price_history_df, *args, **kwargs):
    fig = go.Figure(data=[go.Candlestick(x=price_history_df['Date'],
                    open=price_history_df['Open'],
                    high=price_history_df['High'],
                    low=price_history_df['Low'],
                    close=price_history_df['Close'])])
    
    fig.update_layout(*args, **kwargs)


    fig.show()
    # return fig

make_graph(tesla_data, title='Tesla Price')
```
